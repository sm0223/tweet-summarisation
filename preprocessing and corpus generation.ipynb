# Pre-processing Functions

```python
import string, time, emoji
from langdetect import detect

def remove_html_tags(textArray) :
    res = []
    for text in textArray:
        res.append(re.sub('<.*?>', '', text))
    return res


def remove_urls(textArray) :
    res = []
    for text in textArray:
        z = re.sub('http://\S+|https://\S+', '', text)
        res.append(z)
    return res

def remove_punctuations(textArray) :
    punctuations = ":!”$%&’()*+-/:;<=>[]^_`{|}~•@#"
    res = []
    for text in textArray:
        text = text.translate(str.maketrans('','',punctuations))
        res.append(text)
    return res

def demoji(textArray) :
    res = []
    for text in textArray:
        text = emoji.demojize(text)
        res.append(text)
    return res

def filter_english(textArray):
    res =[]
    for text in textArray:
        try:
            if detect(text) == 'en' :
                res .append(text);
        except:
            continue
    return res
```

# Scraping Data

```python
import snscrape.modules.twitter as sntweet
```

```python
hashtags = ["#IndiaToday",
            "#PakistanEconomy",
            "#SJainshankar",
            "#dedolarisation",
            "#Robinhood",
            "#AyodhyaVerdict",
            "#Iphone15",
            "#FreeTibet",
            "#IndianHistory",
            "#IPL",
            "#crypto",
            "#VandeBharatExpress" 
            ...           
           ]
```

```python
def scrap_tweets(hashtags):
    scrapers = []
    for hashtag in hashtags :
        scraper = sntweet.TwitterSearchScraper(hashtag)
        scrapers.append(scraper)

    topics = []
    for scraper in scrapers:
        tweets = []
        n_tweets = 2500
        for i, tweet in tqdm(enumerate(scraper.get_items()), total = n_tweets):
            data=tweet.content
            tweets.append(data)
            if(i>n_tweets):
                break
        topics.append(tweets)
    return topics
```

```python
data = scrap_tweets(hashtags)
```

```python
data
```

```python
df = []
for topic in data :
    topic = remove_urls(topic)
    topic = filter_english(topic)
    topic = remove_html_tags(topic)
    topic = remove_punctuations(topic)
    topic = demoji(topic)
    df.append(topic)
```

```python
rawList = []
for i, tweets in tqdm(enumerate(data), total = 12):
    for j, tweet in tqdm(enumerate(tweets), total = len(tweets)):  
        x = [j , hashtags[i],tweet]
        rawList.append(x)
pd.DataFrame(rawList, columns = ["index", "tweets", "summary"]).to_csv("rawData.csv")
```
